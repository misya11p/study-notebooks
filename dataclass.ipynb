# Dataclass

データを保持することに特化したクラス。Pythonの標準モジュール`dataclasses`を使って簡単に定義できる。

```python
import dataclasses
```

## 基本的な使い方

例として、こんなクラスを考える。特に機能を持たず、ただデータ（属性）を保持するだけのクラス。

```python
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person("Alice", 30)
print(person.name)
print(person.age)
person
```

Dataclassを使うとこんなクラスを簡単に実装できる。

```python
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int

person = Person("Alice", 30)
print(person.name)
print(person.age)
person
```

## 細かい挙動

```python
@dataclass
class Person:
    name: str
    age: int

person = Person("Alice", 30)
```

存在しない属性にアクセスした場合はエラー

```python
try:
    person.gender
except AttributeError as e:
    print(e)
```

存在しない引数を渡した場合はエラー

```python
try:
    Person("Alice", 30, 0)
except TypeError as e:
    print(e)
```

型チェックは行われない

```python
Person("Alice", "40")
```

デフォルト値の設定が可能

```python
@dataclass
class Person:
    name: str
    age: int = 20

Person("Alice")
```

ミュータブルな属性のデフォルト値を定義する場合は、`field`関数を使う必要がある

```python
from dataclasses import field

@dataclass
class Person:
    name: str
    age: int = 20
    hobbies: list = field(default_factory=list)

Person("Alice")
```

`asdict`で辞書に展開できる

```python
from dataclasses import asdict

asdict(person)
```

dataclassは他のdataclassを属性として持つこともできる。

```python
@dataclass
class Address:
    street: str
    city: str
    zipcode: str

@dataclass
class Person:
    name: str
    age: int
    address: Address

address = Address("123 Main St", "Anytown", "12345")
person = Person("Alice", 30, address)
person
```

```python
print(person.name)
print(person.age)
print(person.address.street)
print(person.address.city)
print(person.address.zipcode)
```

`dacite`というライブラリを使うと、辞書からdataclassのインスタンスを簡単に作成できる。

```python
from dacite import from_dict

args = {
    "name": "Alice",
    "age": 30,
    "address": {
        "street": "123 Main St",
        "city": "Anytown",
        "zipcode": "12345"
    }
}
from_dict(Person, args)
```
